Ejercicio 51:

Encuentra el coeficiente de correlaci칩n entre dos matrices.

```python
import numpy as np 

arr1=np.array([1, 2, 3, 4, 5])
arr2=np.array([3, 4, 5, 6, 7])
correlacion=np.corrcoef(arr1,arr2)[0,1]
print(correlacion)
```

Ejercicio 56:

Crea una matriz aleatoria de 5x5 y reemplaza el valor m치ximo por -1.

```python
import numpy as np

matris=np.random.random((5,5))
maximo=np.max(matris)
matris[matris==maximo]=-1
print(matris)
```
