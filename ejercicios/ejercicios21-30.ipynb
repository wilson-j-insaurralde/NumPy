Ejercicio 21:

Encuentra la media de cada fila en una matriz 2D.

```python
import numpy as np 

matris=np.random.random((3,4))
print(matris)
media=np.mean(matris,axis=1)
print("***"*8)
print(media)
```

Ejercicio 22:

Crea una matriz aleatoria de 4x4 y extrae los elementos diagonales.

```python
import numpy as np 
matris=np.random.random((4,4))
print(matris)
diagonal=np.diag(matris)
print("***"*20)
print(diagonal)
```

Ejercicio 23:

Cuente el número de ocurrencias de un valor específico en una matriz.

```python
import numpy as np
matris=np.array([1, 2, 3, 4, 2, 3, 2, 1])
ocurrencias=np.count_nonzero(matris==2)
print(ocurrencias)
```

Ejercicio 24:

Reemplazar todos los valores en una matriz 1D con la media de la matriz.

```python
import numpy as np
matris=np.array([[10, 20, 30, 40, 50]])
matris[:]=np.mean(matris)
print(matris)
```

Ejercicio 25:

Encuentra los índices de los valores máximo y mínimo en una matriz 1D.

```python
import numpy as np 
matris=np.array([5, 2, 8, 1, 7])
maximo=np.max(matris)
minimo=np.min(matris)

print(maximo)
print(minimo)
```

Ejercicio 26:

Crea una matriz 2D con 1 en el borde y 0 dentro.

```python
import numpy as np 
matris=np.ones((5,5))
matris[1:-1,1:-1]=0
print(matris)
```

Ejercicio 27:

Encuentre los valores únicos y sus conteos en una matriz 1D.

```python
import numpy as np 
matris=np.array([1, 2, 3, 2, 4, 1, 5, 4, 6])
valore_unicos, contador = np.unique(matris, return_counts=True)
print(valore_unicos)
print(contador)
```
