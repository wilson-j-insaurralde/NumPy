Ejercicio 31:

Cree una matriz aleatoria de 3x3 y reemplace todos los valores mayores que 0,5 con 1 y todos los demás con 0.

```python
import numpy as np 
matris=np.random.random((3,3))
print("___"*10)
print(matris)

matris[matris>=0.5]=1
matris[matris<0.5]=0
print("___"*10)
print(matris)
print("___"*10)
```

Ejercicio 32:

Encuentre los índices de los N valores máximos superiores en una matriz 1D.

```python
import numpy as np 
matris=np.array([10, 5, 8, 20, 15])
indices_superiores=np.argsort(matris)[-2:]
print(indices_superiores)
```

Ejercicio 33:

Calcular la media de cada columna en una matriz 2D.

```python
import numpy as np 
matris=np.random.random((4,3))
media=np.mean(matris,axis=0)
print(media)
```

Ejercicio 34:

Normalizar los valores en cada columna de una matriz 2D.

```python
import numpy as np
matris=np.random.random((4,3))
print ("___"*8)
print(matris)
print ("___"*8)
normalizado=(matris-np.mean(matris,axis=0))/np.std(matris,axis=0)
print(normalizado)
```

Ejercicio 37:

Calcular el error cuadrático medio entre dos matrices.

```python
import numpy as np 
mat1=np.array([1,2,3,4])
mat2=np.array([2,3,4,5])
medio=np.mean((mat1-mat2)**2)
print(medio)
```

Ejercicio 39:

Encuentra los percentiles 5 y 95 de una matriz.

```python
import numpy as np
mat1=np.random.random(10)

percentil5=np.percentile(mat1,5)
percentil95=np.percentile(mat1,95)
print(f"matris: {mat1}")
print(f"percentil 5: {percentil5}")
print(f"percentil 95: {percentil95}")
```
