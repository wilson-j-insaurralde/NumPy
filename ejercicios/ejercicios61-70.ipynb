Ejercicio 61:

Calcular el percentil a lo largo de un eje específico de una matriz 2D.

```python
import numpy as np 

matris=np.random.random((3,2))
percentil=np.percentile(matris,75,axis=1)
print(percentil)
```

Ejercicio 62:

Crea una matriz 1D y agrega un borde de 0 a su alrededor.

```python
import numpy as np 
matris=np.array([1,2,3,4])
borde=np.pad(matris,(1,1), mode='constant',constant_values=0)
print(borde)
```

Ejercicio 63:

Calcular el histograma de una matriz 1D.

```python
import numpy as np

matris=np.array([1, 1, 2, 2, 2, 3, 3, 3, 3])
hist, bins = np.histogram(matris, bins=[1, 2, 3, 4])
print("histograma:", hist)
print("bordes:", bins)
```

Ejercicio 64:

Crea una matriz 2D con valores aleatorios y normaliza cada fila.

```python
import numpy as np 
matris=np.random.random((4,3))
normalizada = matris / np.linalg.norm(matris, axis=1, keepdims=True)
print(normalizada)
```

```python
import numpy as np 

matris=np.random.random((3,4))
sorted_matrix_by_column2 = matris[matris[:, 1].argsort()]
print(sorted_matrix_by_column2)

print(" ##### "*8)
print(matris)
```

Ejercicio 69:

Crea una matriz 2D con valores aleatorios y reemplaza el valor máximo por el mínimo.

```python
import numpy as np 
matris=np.random.random((3,4))
max_value_index = np.unravel_index(np.argmax(matris), matris.shape)
min_value = np.min(matris)
matris[max_value_index] = min_value
print(matris)
```
