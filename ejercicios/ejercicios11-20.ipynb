Invertir una matriz 1D (el primer elemento se convierte en el último).

```python
import numpy as np 
arr=np.arange(10)
invertir=np.flip(arr)
print(arr)
print("___"*8)
print(invertir)
```

Ejercicio 15:

Obtener los elementos comunes entre dos matrices.

```python
import numpy as np 
matris1=np.array([1,2,3,4,5])
matris2=np.array([3,4,5,6,7])
igual=np.intersect1d(matris1,matris2)
print(igual)
```

Ejercicio 18:

Normaliza los valores en una matriz 1D entre 0 y 1.

```python
import numpy as np 
tuki=np.array([2, 5, 10, 3, 8])
areglo=(tuki-np.min(tuki))/(np.max(tuki)-np.min(tuki))
print(areglo)
min=np.min(tuki)
max=np.max(tuki)
print(min,max)
```
