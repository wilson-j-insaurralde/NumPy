Calcular la suma de los elementos diagonales en una matriz 2D.

```python
import numpy as np
matrix = np.random.random((3, 3))
diagonal_sum = np.trace(matrix)
print(diagonal_sum)
```

Ejercicio 92:

Convierte radianes a grados para cada elemento de una matriz 1D.

```python
import numpy as np
arr_in_radians = np.array([np.pi/2, np.pi, 3*np.pi/2])
arr_in_degrees = np.degrees(arr_in_radians)
print(arr_in_degrees)
```

Ejercicio 93:

Calcular la distancia euclidiana entre dos matrices.

```python
import numpy as np
arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])
euclidean_distance = np.linalg.norm(arr1 - arr2)
print(euclidean_distance)
```

Ejercicio 94:

Cree una matriz 1D y establezca los valores entre el percentil 25 y 75 en 0.

```python
import numpy as np
arr = np.array([10, 20, 30, 40, 50])
percentile_25th = np.percentile(arr, 25)
percentile_75th = np.percentile(arr, 75)
arr[(arr >= percentile_25th) & (arr <= percentile_75th)] = 0
print(arr)
```

Ejercicio 96:

Reemplaza todos los números pares en una matriz 1D con el siguiente número impar.

```python
import numpy as np
arr = np.array([2, 5, 8, 12, 15])
arr[arr % 2 == 0] += 1
print(arr)
```

Ejercicio 97:

Crea una matriz 2D y normaliza cada columna por su rango.

```python
import numpy as np
matrix = np.random.random((3, 4))
normalized_columns_range = (matrix - np.min(matrix, axis=0)) / (np.max(matrix, axis=0) - np.min(matrix, axis=0))
print(normalized_columns_range)
```


Ejercicio 98:

Calcular la suma acumulada de elementos a lo largo de un eje dado en una matriz 2D.

```python
import numpy as np
matrix = np.random.random((3, 4))
cumulative_sum_axis1 = np.cumsum(matrix, axis=1)
print(cumulative_sum_axis1)
```

Ejercicio 99:

Comprueba si algún elemento de una matriz 1D es distinto de cero.

```python
import numpy as np
arr = np.array([0, 0, 0, 1, 0])
any_nonzero = np.any(arr != 0)
print(any_nonzero)
```
