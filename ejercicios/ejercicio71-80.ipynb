Ejercicio 71:

Cree una matriz 1D y establezca los valores entre 10 y 20 en 0.

```python
import numpy as np

arr=np.array([5, 15, 12, 18, 25])
arr[(arr>=10) & (arr<=20)]=0
print(arr)

```

Ejercicio 73:

Calcular el producto Kronecker de dos matrices.

```python
import numpy as np 

arr1=np.array([1,2])
arr2=np.array([3,4])
pro=np.kron(arr1,arr2)
print(pro)
```
