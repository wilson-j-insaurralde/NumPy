Ejercicio 71:

Cree una matriz 1D y establezca los valores entre 10 y 20 en 0.

```python
import numpy as np

arr=np.array([5, 15, 12, 18, 25])
arr[(arr>=10) & (arr<=20)]=0
print(arr)

```

Ejercicio 73:

Calcular el producto Kronecker de dos matrices.

```python
import numpy as np 

arr1=np.array([1,2])
arr2=np.array([3,4])
pro=np.kron(arr1,arr2)
print(pro)
```

Ejercicio 75:

Crea una matriz de 3x3 y establece todos los valores por encima de la diagonal principal en cero.

```python
import numpy as np 

matris=np.random.random((3,3))
matris[np.triu_indices(3, 1)] = 0
print(matris)
```

Ejercicio 76:

Cuente el número de ocurrencias de cada valor único en una matriz 1D.

```python
import numpy as np
arr = np.array([2, 2, 1, 3, 3, 3, 4])
unique_values, counts = np.unique(arr, return_counts=True)
print("Unique values:", unique_values)
print("Counts:", counts)
```

Ejercicio 77:

Calcular el producto acumulativo de elementos a lo largo de un eje dado en una matriz 2D.

```python
import numpy as np
matrix = np.random.random((3, 4))
cumulative_product_axis0 = np.cumprod(matrix, axis=0)
print(cumulative_product_axis0)
```
