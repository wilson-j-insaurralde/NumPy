NumPy ofrece el subm√≥dulo numpy.random, con el que podemos generar n√∫meros aleatorios simples, arrays completos y muestras de distribuciones estad√≠sticas muy utilizadas.

9.1 Generaci√≥n de n√∫meros aleatorios (np.random)
Existen m√∫ltiples formas de generar n√∫meros aleatorios:

a) Un n√∫mero aleatorio entre 0 y 1 (random.rand)

```python
import numpy as np

print("N√∫mero aleatorio [0,1):", np.random.rand())
```

b) Array de n√∫meros aleatorios (rand)

```python
print("Array 1D:", np.random.rand(5))        # 5 n√∫meros aleatorios
print("Matriz 2D:\n", np.random.rand(2, 3))  # Matriz 2x3
```

c) Enteros aleatorios (randint)

```python
print("Enteros aleatorios:", np.random.randint(1, 10, size=5))  
# 5 n√∫meros entre 1 y 9
```

9.2 Distribuciones comunes
NumPy permite simular distribuciones estad√≠sticas.

a) Distribuci√≥n uniforme (np.random.uniform)
Genera valores entre un rango [a, b) con probabilidad uniforme.

```python
arr = np.random.uniform(0, 10, size=5)
print("Distribuci√≥n uniforme [0,10):", arr)
```

b) Distribuci√≥n normal o gaussiana (np.random.normal)
Muy utilizada en estad√≠stica y machine learning.

```python
arr = np.random.normal(loc=0, scale=1, size=10)
print("Distribuci√≥n normal (media=0, std=1):", arr)
```

üëâ loc es la media; scale, la desviaci√≥n est√°ndar.

c) Distribuci√≥n binomial (np.random.binomial)
Se usa para experimentos ‚Äús√≠/no‚Äù (Bernoulli).

```python
# Simulaci√≥n: 10 experimentos, probabilidad de √©xito 0.5, repetido 5 veces
arr = np.random.binomial(n=10, p=0.5, size=5)
print("Distribuci√≥n binomial:", arr)
```

üëâ Ejemplo t√≠pico: lanzar una moneda 10 veces y contar cu√°ntas veces sali√≥ cara.

9.3 Fijar semillas (np.random.seed)
Sin semilla, cada ejecuci√≥n produce valores distintos. Para reproducir resultados, fijamos una semilla.

```python
np.random.seed(42)

print("Con semilla fija:", np.random.randint(1, 100, 5))
```

Salida (repetible):

Con semilla fija: [52 93 15 72 61]
üëâ Es clave para experimentar y comparar modelos de forma justa.

Resumen
- np.random.rand(): n√∫meros aleatorios en [0,1).
- np.random.randint(a, b, size): enteros aleatorios.
- np.random.uniform(a, b, size): distribuci√≥n uniforme.
- np.random.normal(media, std, size): distribuci√≥n normal.
- np.random.binomial(n, p, size): distribuci√≥n binomial.
- np.random.seed(valor): fija la semilla para reproducibilidad.
Nota: Estas herramientas permiten crear simulaciones, experimentos y pruebas controladas en proyectos de ciencia de datos.
