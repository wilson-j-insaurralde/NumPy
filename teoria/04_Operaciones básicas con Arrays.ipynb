4.1 Operaciones aritm칠ticas elemento a elemento
Podemos aplicar directamente operaciones aritm칠ticas (+, -, *, /, **) sobre arrays; NumPy las ejecuta elemento por elemento.

```python
import numpy as np

a = np.array([1, 2, 3, 4])
b = np.array([10, 20, 30, 40])

print("Suma:", a + b)           # [11 22 33 44]
print("Resta:", b - a)         # [ 9 18 27 36]
print("Multiplicaci칩n:", a * b)  # [ 10  40  90 160]
print("Divisi칩n:", b / a)     # [10. 10. 10. 10.]
print("Potencia:", a ** 2)      # [ 1  4  9 16]
```

4.2 Broadcasting en NumPy
El broadcasting permite operar entre arrays de diferentes tama침os o formas, siempre que se cumplan reglas espec칤ficas.

Ejemplo 1: sumar un escalar a un array.

```python
arr = np.array([1, 2, 3, 4])
print(arr + 5)  # [6 7 8 9]
```

Ejemplo 2: operaciones entre un array 2D y uno 1D.

```python
mat = np.array([[1, 2, 3],
                [4, 5, 6]])

vec = np.array([10, 20, 30])

print(mat + vec)
```

```python
arr = np.array([1, 4, 9, 16])
print("Ra칤z cuadrada:", np.sqrt(arr))  # [1. 2. 3. 4.]
```

c) Logaritmo natural (np.log)

```python
arr = np.array([1, np.e, np.e**2])
print("Logaritmo natural:", np.log(arr))  # [0. 1. 2.]
```

d) Funciones trigonom칠tricas (np.sin, np.cos)
