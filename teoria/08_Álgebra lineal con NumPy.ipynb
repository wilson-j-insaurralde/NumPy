El √°lgebra lineal es esencial para representar y manipular datos en m√∫ltiples dimensiones: vectores, matrices y tensores. NumPy incluye el subm√≥dulo numpy.linalg con operaciones optimizadas en C y Fortran, ideales para c√°lculos r√°pidos.

8.1 Producto punto (dot)
El producto punto puede calcularse entre vectores o matrices compatibles.

```python
import numpy as np

# Producto punto entre vectores
v1 = np.array([1, 2, 3])
v2 = np.array([4, 5, 6])

print("Producto punto (vectores):", np.dot(v1, v2))  # 32
# 1*4 + 2*5 + 3*6 = 32
```

üëâ Tambi√©n funciona entre matrices:

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])

print("Producto punto (matrices):\n", np.dot(A, B))
# [[1*5 + 2*7, 1*6 + 2*8],
#  [3*5 + 4*7, 3*6 + 4*8]]
```

Salida:

[[19 22]
 [43 50]]
8.2 Multiplicaci√≥n de matrices
NumPy ofrece varias formas de multiplicar matrices:

np.dot(A, B): producto matricial.
A @ B: operador legible (Python 3.5+).
np.matmul(A, B): equivalente, pensado para matrices.

```python
print("Usando @:\n", A @ B)
print("Usando matmul:\n", np.matmul(A, B))
```

üëâ La multiplicaci√≥n elemento a elemento se realiza con *, no con dot:

```python
print("Multiplicaci√≥n elemento a elemento:\n", A * B)
```

8.3 Matrices identidad e inversas
a) Matriz identidad
La matriz identidad cumple el rol del ‚Äú1‚Äù en √°lgebra lineal.

```python
I = np.eye(3)
print("Matriz identidad 3x3:\n", I)
```

Salida:

[[1. 0. 0.]
 [0. 1. 0.]
 [0. 0. 1.]]
b) Inversa de una matriz
La matriz inversa A‚Åª¬π (si existe) satisface A @ A‚Åª¬π = I.

```python
A = np.array([[1, 2],
              [3, 4]])

A_inv = np.linalg.inv(A)
print("Matriz inversa:\n", A_inv)

# Verificar
print("A @ A_inv:\n", A @ A_inv)
```

Salida (aprox.):

Matriz inversa:
 [[-2.   1. ]
  [ 1.5 -0.5]]

A @ A_inv:
 [[1. 0.]
  [0. 1.]]

8.4 Determinantes y autovalores
a) Determinante (np.linalg.det)
Permite saber si una matriz es invertible y aparece en m√∫ltiples c√°lculos.

```python
A = np.array([[1, 2],
              [3, 4]])

det = np.linalg.det(A)
print("Determinante:", det)  # -2.0...
```

b) Autovalores y autovectores (np.linalg.eig)
Se utilizan en reducci√≥n de dimensionalidad (PCA) y modelos matem√°ticos.

```python
A = np.array([[2, 0],
              [0, 3]])

valores, vectores = np.linalg.eig(A)

print("Autovalores:", valores)
print("Autovectores:\n", vectores)
```

Salida:

Autovalores: [2. 3.]
Autovectores:
 [[1. 0.]
  [0. 1.]]
üëâ Como la matriz es diagonal, los autovalores coinciden con sus elementos principales.

Resumen
- np.dot() o @: producto punto y multiplicaci√≥n matricial.
- np.eye(n): matriz identidad.
- np.linalg.inv(A): inversa de una matriz.
- np.linalg.det(A): determinante.
- np.linalg.eig(A): autovalores y autovectores.
Nota: Estas operaciones son esenciales en procesamiento de datos, Machine Learning y Deep Learning.
