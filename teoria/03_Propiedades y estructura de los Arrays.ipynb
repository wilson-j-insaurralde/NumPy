3.1 Dimensiones, forma, tama침o y tipo de datos
a) Dimensiones (ndim)
Indica cu치ntas dimensiones tiene el array:

1D: vector.
2D: matriz.
3D o m치s: tensores.

```python
import numpy as np

a = np.array([1, 2, 3])                 # 1D
b = np.array([[1, 2, 3], [4, 5, 6]])    # 2D
c = np.array([[[1], [2]], [[3], [4]]])  # 3D

print("a.ndim:", a.ndim)  # 1
print("b.ndim:", b.ndim)  # 2
print("c.ndim:", c.ndim)  # 3
```

b) Forma (shape)
Devuelve una tupla con la cantidad de elementos por dimensi칩n.

```python
print("Forma de b:", b.shape)  # (2, 3) -> 2 filas y 3 columnas
```

c) Tama침o (size)
Indica la cantidad total de elementos del array.

```python
print("Tama침o de b:", b.size)  # 6 elementos en total
```

d) Tipo de datos (dtype)
Los arrays son homog칠neos: todos los elementos comparten el mismo tipo.

```python
arr = np.array([1, 2, 3])
print("Tipo de datos:", arr.dtype)  # int64 (depende del sistema)
```
