Una de las principales fortalezas de NumPy es la facilidad y flexibilidad para crear arrays. Podemos generarlos a partir de estructuras de Python (listas o tuplas) o con funciones propias de la biblioteca que producen arreglos de forma autom치tica.

2.1 Crear arrays desde listas o tuplas
Podemos convertir listas o tuplas de Python en arrays de NumPy usando np.array().

```python
import numpy as np
# Desde una lista
lista = [1, 2, 3, 4, 5]
arr1 = np.array(lista)
print("Array desde lista:", arr1)
# Desde una tupla
tupla = (10, 20, 30, 40)
arr2 = np.array(tupla)
print("Array desde tupla:", arr2)
# Array 2D (matriz)
lista2d = [[1, 2, 3], [4, 5, 6]]
arr3 = np.array(lista2d)
print("Array 2D:\n", arr3)
```

游녤 Not치 que NumPy infiri칩 autom치ticamente que todos los elementos son enteros.

2.2 Funciones para generar arrays
Adem치s de convertir listas, NumPy ofrece funciones para crear arrays de manera r치pida.

a) np.arange(start, stop, step)
Genera un array con valores en un rango definido, similar a range() de Python, pero devuelve un array.

```python
arr = np.arange(0, 10, 2)
print("np.arange:", arr)
```

b) np.linspace(start, stop, num)

Genera un array con una cantidad fija de elementos, distribuidos de forma equidistante entre un valor inicial y uno final.

```python
arr = np.linspace(0, 1, 5)
print("np.linspace:", arr)
```

游녤 Ideal cuando quer칠s dividir un intervalo en n puntos iguales (muy usado en gr치ficas y simulaciones).

c) np.zeros(shape)
Crea un array lleno de ceros.

```python
arr = np.zeros((2, 3))  # 2 filas, 3 columnas
print("np.zeros:\n", arr)
```

d) np.ones(shape)
Crea un array lleno de unos.

```python
arr = np.ones((3, 3))
print("np.ones:\n", arr)
```

e) np.eye(n)
Genera una matriz identidad de tama침o n 칑 n, con unos en la diagonal principal y ceros en el resto.

```python
arr = np.eye(4)
print("np.eye:\n", arr)
```

游녤 Muy usada en 치lgebra lineal.

Resumen
- np.array(lista/tupla): convierte estructuras de Python en arrays.
- np.arange(inicio, fin, paso): secuencia con paso fijo.
- np.linspace(inicio, fin, cantidad): secuencia con cantidad fija de valores.
- np.zeros(shape): array de ceros.
- np.ones(shape): array de unos.
- np.eye(n): matriz identidad.
